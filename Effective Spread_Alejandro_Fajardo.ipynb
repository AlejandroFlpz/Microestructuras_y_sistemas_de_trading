### Paqueterias

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Funciones

```python
def Serial_covariance(df, windows, precios='Close'):
    
    df['Diff'] = df[precios].diff()
    df['Diff-1'] = df['Diff'].shift(-1)
    df = df.dropna().reset_index(drop=True)
    
    for i in windows:
        covarianza = df['Diff'].rolling(window=i).cov(df['Diff-1'])
        df[f'Spread_{i}'] = 2 * np.sqrt(np.abs(covarianza))
        
        df[f'Bid_{i}'] = np.where(df['Diff'] > 0, df[precios] - df[f'Spread_{i}'], df[precios])
        df[f'Ask_{i}'] = np.where(df['Diff'] > 0,df[precios], df[precios] + df[f'Spread_{i}'])
    
    return df

def block_method(df, windows, precios='Close'):
    df['Diff'] = df[precios].diff()
    df = df.dropna().reset_index(drop=True)


    for t in windows:
        spreads = [np.nan] * len(df)

        for i in range(len(df) - 2*t + 1):
            W1 = df['Diff'].iloc[i:i+t]
            W2 = df['Diff'].iloc[i+t:i+2*t]
            covarianzas = np.cov(W1, W2)[0,1]
            spreads[i + 2*t - 1] = 2 * np.sqrt(abs(covarianzas))

        df[f'Spread_{t}'] = spreads

        df[f'Bid_{t}'] = np.where(df['Diff'] > 0, df[precios] - df[f'Spread_{t}'], df[precios])
        df[f'Ask_{t}'] = np.where(df['Diff'] > 0, df[precios], df[precios] + df[f'Spread_{t}'])

    return df
```

### Limpieza de datos

```python
df = pd.read_csv('aapl_5m_train.csv')
df = df.drop(['Unnamed: 0','Timestamp','Gmtoffset', 'Datetime', 'Open', 'High', 'Low', 'Volume' ], axis=1)
df_S = df.copy()
df_B = df.copy()
windows = [5, 10, 20, 50, 100]
```

### Serial Covariance

```python
df_S = Serial_covariance(df_S, windows)
df_S
```

### Block Method

```python
df_B = block_method(df_B, windows)
df_B
```

### Average Spreads AAPL

```python
avg_spread_per_window = pd.DataFrame({
    'Window': windows,
    'Avg Spread S_Cov':[df_S[f'Spread_{w}'].mean() for w in windows],
    'Avg Spread B_Method': [df_B[f'Spread_{w}'].mean() for w in windows]
})

avg_spread_per_window
```

#### Gráfica

```python
windows = avg_spread_per_window['Window']
s_cov = avg_spread_per_window['Avg Spread S_Cov']
b_method = avg_spread_per_window['Avg Spread B_Method']

x = np.arange(len(windows))
width = 0.35  

plt.bar(x - width/2, s_cov, width, label='S_Cov')
plt.bar(x + width/2, b_method, width, label='B_Method')


plt.xticks(x, windows)
plt.xlabel('Windows')
plt.ylabel('Average Spread')
plt.title('Average Spread por window')
plt.legend()
plt.show()
```

```python
spread_S = [c for c in df_S.columns if "Spread" in c]
df_S[spread_S].plot(figsize=(12,6), title="Spreads Serial Covariance", lw=1)
plt.xlabel("Tiempo")
plt.ylabel("Spread")
plt.show()
```

```python
spread_B = [cl for cl in df_B.columns if "Spread" in cl]
df_B[spread_B].plot(figsize=(12,6), title="Spreads Block Method", lw=1)
plt.xlabel("Tiempo")
plt.ylabel("Spread")
plt.show()
```

### Analísis

Tras haber obtenido el calculo de los spreads mediante dos métodos distintos como lo puede ser el metodo de serial covariance así como el de bloque, se pueden analizar distintas situaciones. 

Primero podemos observar como es que van cambiando los spreads conforme se van haciendo más grandes las ventanas. Tanto en el método de serial covariance como en el del bloque, se puede ver la misma tendencia, donde mientras menor sea el número de la ventana, mayor va a ser el spread en promedio obtenido. Esto quiere decir que, los spreads conforme las ventanas son mayores, estos tienen una mayor estabilización. De igual forma, es posible ver en ambos casos que, aunque el spread vaya bajando cuando las ventanas se hacen más grandes, en todas las ventanas el spread tiende a ser muy bajo, mostrandonos que el activo tiene una alta liquidez. A pesar de que ambos reflejan la misma tendencia, si es visible una ligera diferencia entre los dos métodos. En el caso de Block, tiene un spread más alto en las ventanas de 5 y 10, mientras que este es más bajo en las ventanas de 20, 50 y 100. 

Como en este caso, estamos buscando inferir que tipo de operación fue la ultima que se hizo, el mejor método para realizar esto efectivamente es el de Serial covariance. En este método se van comparando ventanas donde una de ellas solo tiene rezago de un dato, entonces la serie puede tener un mejor ajuste por la comparación continua de los precios y el cálculo de sus covarianzas que se permite en este caso. Por otro lado, en el Block Method, como las ventanas y las comparaciones son por bloques, se agrupan más las covarianzas y los cambios en el tiempo no suelen capturarse bien, provocando que la deducción de saber si la ultima transacción fue un bid o ask sea más compleja.

Se puede concluir, que en nuestro caso como se busca inferir el tipo de operaciones que se fueron realizando, el mejor método es el de serial covariance, pemitiendonos saber de mejor forma si fue un bid o ask el ultimo movimiento. Asimismo, sabemos que este método al tener spread muy bajos, nos ayuda a tener un calculo más certero sobre el spread efectivo, sin arriesgar la posibilidad de inferir el tipo de operación realizada.
